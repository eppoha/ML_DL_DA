除了删除缺失值，对缺失值处理的另外一种方法就是填充缺失值。
一，手动填充。
二，临近填充。
三，插值填充。


```python
# 用前面的临近值进行填充
df.fillna(method='pad')
```

```python
# 用后面的临近值进行填充
df.fillna(method='bfill')
```

```python
# 线性插值填充。
df.interpolate()
```
